```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import joblib

data = pd.read_csv('train_data.csv')

# Признаки
features = ['veh_1', 'veh_2', 'veh_3', 'veh_5', 'veh_10', 'veh_15', 'veh_20', 'veh_30', 'hour']
X = data[features]

# Для каждого горизонта своя модель
for horizon, target in [('30', 'target_30'), ('60', 'target_60'), ('120', 'target_120')]:
    y = data[target]
    
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X, y)
    
    joblib.dump(model, f'model_{horizon}.pkl')
    print(f"Модель для {horizon} мин сохранена.")
```
