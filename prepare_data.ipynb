```python
import pandas as pd


df = pd.read_csv('historical_combined.csv')


transport = ['car', 'truck', 'bus', 'motorcycle']
df = df[df['class'].isin(transport)]

# Добавляем минуту
df['minute'] = df['time_ms'] // 60000

# Группируем по минутам: сколько уникальных машин
minute_stats = df.groupby('minute')['track_id'].nunique().reset_index()
minute_stats.columns = ['minute', 'vehicles']


for lag in [1, 2, 3, 5, 10, 15, 20, 30]:
    minute_stats[f'veh_{lag}'] = minute_stats['vehicles'].shift(lag)


minute_stats['hour'] = minute_stats['minute'] % 24


minute_stats['target_30'] = minute_stats['vehicles'].shift(-30)
minute_stats['target_60'] = minute_stats['vehicles'].shift(-60)
minute_stats['target_120'] = minute_stats['vehicles'].shift(-120)


minute_stats = minute_stats.dropna()

# Сохраняем
minute_stats.to_csv('train_data.csv', index=False)
print(f"Создано {len(minute_stats)} строк для обучения.")
```
